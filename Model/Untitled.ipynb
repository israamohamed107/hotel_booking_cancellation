```python
from sklearn.utils import resample
```

```python
not_canceled_dowenSample=resample(not_canceled,
                                replace = False, # sample without replacement
                                n_samples = len(canceled), # match minority n
                                random_state = 42)
not_canceled_dowenSample
```

```python
dowen_sampled=pd.concat([not_canceled_dowenSample,canceled])
dowen_sampled
```

```python
x_trian=dowen_sampled.drop('is_canceled',axis=1)
y_train=dowen_sampled['is_canceled']
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix , accuracy_score,classification_report
```

```python
models={
    'LogisticRegression': LogisticRegression(),
    'Naive Bayes':GaussianNB(),
    'RandomForest': RandomForestClassifier(n_estimators=10),
    'Decision Tree': DecisionTreeClassifier(),
}
```

```python
from sklearn.feature_selection import SelectFromModel
```

```python
from sklearn.linear_model import Lasso
```

```python
select=SelectFromModel(Lasso(alpha=0.005, random_state=42))
```

```python
select.fit(x,y)
```

```python
select.get_support()
```

```python
x.shape[1] # columns
```

```python
cols=x.columns
cols
```

```python
selected_features=cols[select.get_support()]
selected_features
```

```python
len(selected_features)
```

```python
x_trian=x_trian[selected_features]
x_test=x_test[selected_features]
```

```python
x_trian=scaler.fit_transform(x_trian)
x_test=scaler.transform(x_test)
```

```python
x_trian.shape
```

```python
for name, model in models.items():
    print(name)
    model.fit(x_trian, y_train)
    
    # Make predictions.
    y_pred= model.predict(x_test)

    # Compute the error.
    print('Trainning score',model.score(x_trian,y_train))
    print(confusion_matrix(y_pred, y_test))
    print(accuracy_score(y_pred,y_test))
    print('\n')
```
